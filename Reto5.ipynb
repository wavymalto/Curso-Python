## Ejercicio 1: Funciones


De nuestro código que evaluaba cual era una el número mayor, convertirla a función.

Entrada: num1, num2

Salida: 

1. El num1 es mayor que num2
2. El num2 es mayor que num2
3. Ambos son iguales

````python
def mayorQue(numero_1,numero_2):
    """
    FUncion que evalua numeros.
    """
    if numero_1>numero_2:
        resultado= print('num1 es mayor que num2')
    elif numero_2 > numero_1:
        resultado= print('num2 es mayor que num1')
    else: 
        resultado= print('Son iguales')
    return resultado


    
````

````python
x=int(input('Dame un valor 1'))
y=int(input('Dame un valor 2'))
mayorQue(x,y)
````

## Ejercicio 2:

Crear una función que reciba como parameto un texto y que devuelva en el `return` una lista que contenga cada palabra del texto.

|Entrada | Salida |
| - | - | 
| "El señor de los anillos" | ["El", "señor", "de", "los", "anillos"] |



Utiliar el metodo .split() al texto

**Ejemplo**

"Buenos días".split()

````python
def textoALista(texto):
    """
    Funcion que vuelve un texto a una  lista que contenga cada palabra del texto.
    """
    lista= texto.split()
    return lista

texto="El señor de los anillos" 
textoALista(texto)
````

## Ejercicio 3:

Un piloto de carreras ha registrado su velocidad cada segundo en la siguiente lista:

velocidad = [0, 60, 80, 117, 140, 155, 170, 189, 205, 223, 234, 239, 234, 227, 215, 180, 155, 160, 189, 206, 235]

Hacer una función que imprima en que segundo el piloto rebaso los 210 km/h


**Ejemplo de impresión**

En el segundo `9` el piloto llego a los `223` km/h




````python
def limiteVelocidad(velocidades):
    segundos=0
    for velocidad in velocidades:
        if velocidad > 210:
            print(f'En el segundo {segundos} el piloto {velocidad}')
        segundos= segundos+1
````

````python
vel = [0, 60, 80, 117, 140, 155, 170, 189, 205, 223, 234, 239, 234, 227, 215, 180, 155, 160, 189, 206, 235]
limiteVelocidad(vel)
````

## Extra points

Hacer una función que nos imprima cuantos números pares e impares tiene una lista.

Para evaluar un par o impar puedes utilizar el

```Python
if numero%2 = 0:
    print("numero es par")
else:
    print("numero es impar")
```

|Entrada | Salida |
| - | - | 
| [2,3,6,8,9] | Tiene `3` pares y `2` inpares |

````python
def numeroParesInpares(lista):
    contador_par=0
    contador_inpar=0
    for numero in lista:
        if numero%2 ==0:
            contador_par+=1
        else:
            contador_inpar+=1
    return f'Tiene {contador_par} pares y {contador_inpar} impares'
````

````python
numeros=[2,3,6,8,9]
numeroParesInpares(numeros)
````
