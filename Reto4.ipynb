## Ejercicio 3: For con input

Crea un programa que pregunte **5** veces las calificaciones de un alumno, las almacene en una lista vacia y cuando termine el ciclo, devolver el promedio de las calificaciones

```python
n = int(input("Dame 5 calificaciones"))

calificaciones = []
for i in range(n):
    calificacion =int(input(f"Dame la calificacion {i+1}"))
    calificaciones.append(calificacion)
promedio = sum(calificaciones) / len(calificaciones)

if promedio >=6:
    print(f"Felicidades obtuviste {promedio} y aprobaste el semestre")
else:
    print("F")
```
